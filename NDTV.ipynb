```python
import requests
from bs4 import BeautifulSoup
li =[]
url = "https://archives.ndtv.com/articles/2023-01.html"

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, 'html.parser')

    # Find the specific 'ul' tag with the given class
    ul_tag = soup.find('div',  id='main-content')

    # Check if the 'ul' tag is found
    if ul_tag:
        # Find all 'a' (anchor) tags within the 'ul' tag
        links = ul_tag.find_all('a')

        # Extract and print the href attribute of each link
        for link in links:
            href = link.get('href')
            if href:
                li.append(href)
                print(href)
    else:
        print("Could not find the specified 'ul' tag with class.")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import re
from datetime import datetime

URL_list = li[1:315]
URL_list2 = li[316:920]
URL_list3 = li[921:1620]

# Function to scrape information from a given URL
def scrape_info(url):
    # Fetch HTML content
    response = requests.get(url)
    html_code = response.text
    
    # Parse the HTML content
    soup = BeautifulSoup(html_code, 'html.parser')
    
    editor_names = []

    # Find all elements with itemprop="author"
    author_elements = soup.find_all('span', itemprop='author')

    # Iterate through each author element and collect the names
    for author_element in author_elements:
        name_element = author_element.find('meta', itemprop='name')
        if name_element:
            editor_names.append(name_element['content'])
    
    editor_name = ','.join(editor_names)
    
    # Extract the information
    editor_name_element = soup.select_one('span[itemprop="name"]') or soup.select_one('meta[itemprop="name"]')
    editor_name_2 = editor_name_element.get('content') if editor_name_element else None
    
    published_date = soup.select_one('.pst-by_li:contains("Updated:")')
    published_date = published_date.text.strip() if published_date else None
    
    language = soup.select_one('meta[itemprop="inLanguage"]')['content'] if soup.select_one('meta[itemprop="inLanguage"]') else None
    publisher_name = soup.select_one('meta[itemprop="name"]')['content'] if soup.select_one('meta[itemprop="name"]') else None
    
    return {
        'URL' : url,
        'editor_name': editor_name,
        'editor_2nd' : editor_name_2,
        'published_date': published_date,
        'language': language,
        'publisher_name': publisher_name
    }


data_list = []
a = []
# Iterate through the list of URLs
for url in URL_list2:
    info = scrape_info(url)
    data_list.append(info)
    
    # Check if 'published_date' is not None before applying regex
    if info['published_date'] is not None:
        # Use regular expression to find the numeric part
        matches = re.findall(r'\b\d+\b', info['published_date'])
        
        # Extract the first match (assuming there is at least one numeric part in the string)
        day = int(matches[0]) if matches else None
        
        # Create a datetime object with January as the month and 2023 as the year
        date_object = datetime(year=2023, month=1, day=day) if day else None
        
        # Format the date as "x/01/2023"
        formatted_date = date_object.strftime("%d/%m/%Y") if date_object else None
    else:
        formatted_date = None
    
    # appending in a list
    a.append(formatted_date)


# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data_list)

# Add the 'formatted_date' list to the DataFrame
df['Formatted Date'] = a

# Save the DataFrame to an Excel file
df.to_excel('scraped_data_3JAN.xlsx', index=False)
```
